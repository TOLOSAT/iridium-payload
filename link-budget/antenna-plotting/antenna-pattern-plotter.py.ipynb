```python
import numpy as np
import matplotlib.pyplot as plt


# useful module for writting pathes
from pathlib import Path
```

```python
# Initialize variables
data_folder = Path(".")
file_to_open = data_folder / "iri_011_001.txt"
f = open(file_to_open)
l = []

# Read the file
l= f.readlines()

# Select interesting data in the file
l=l[17:]
#print(l)

l_splitted=[]
for elt in l:
    l_splitted.append(elt.split())


#print(l_splitted)
theta, rho = np.linspace(0, 2*np.pi, 361), []

for elt in l_splitted:
    #theta.append(elt[0])
    rho.append(elt[1])
print(theta)
#print(max(rho))

rho = [10**float(elt) for elt in rho]
#print(rho)

# Plot
plt.figure()
ax = plt.subplot(111, polar=True)
ax.plot(theta, rho ,color='r')
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
#ax.set_rmin(10**(min(rho)))
#ax.set_rmax(10**(max(rho)))
ax.set_title("Rayonnement de l'antenne Maxtenna MPA-D254-1621", va='bottom')
#theta = np.linspace(0.01,2*np.pi,500)
plt.tight_layout()
#plt.savefig("dipole1.png"); plt.savefig("dipole1.pdf")
plt.show()
```
